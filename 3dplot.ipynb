```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Adjusting the axis limits to ensure uniform scaling across all dimensions
def draw_sphere(ax, x, y, z, diameter):
    """Draw a sphere in 3D space at (x, y, z) with the given diameter."""
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    r = diameter / 2
    X = r * np.outer(np.cos(u), np.sin(v)) + x
    Y = r * np.outer(np.sin(u), np.sin(v)) + y
    Z = r * np.outer(np.ones(np.size(u)), np.cos(v)) + z
    ax.plot_surface(X, Y, Z, color='b', alpha=0.6)

def draw_point(ax, x, y, z):
    """Draw a point in 3D space at (x, y, z)."""
    ax.scatter(x, y, z, color='r', s=10)
    
def draw_line(ax, x1, y1, z1, x2, y2, z2):
    """Draw a line in 3D space from (x1, y1, z1) to (x2, y2, z2)."""
    ax.plot([x1, x2], [y1, y2], [z1, z2], color='g')
    
# Create a new figure for 3D plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw the sphere and point again
draw_sphere(ax, 30, 50, 0, 80)
draw_point(ax, 0, 0, 0)
draw_point(ax, 0, 100, 0)
draw_line(ax, 0, 0, 0, 0, 100, 0)

# Setting the same range for all axes to maintain aspect ratio
max_range = np.array([ax.get_xlim(), ax.get_ylim(), ax.get_zlim()]).ptp().max() / 2

mid_x = (ax.get_xlim()[0] + ax.get_xlim()[1]) / 2
mid_y = (ax.get_ylim()[0] + ax.get_ylim()[1]) / 2
mid_z = (ax.get_zlim()[0] + ax.get_zlim()[1]) / 2

ax.set_xlim(mid_x - max_range, mid_x + max_range)
ax.set_ylim(mid_y - max_range, mid_y + max_range)
ax.set_zlim(mid_z - max_range, mid_z + max_range)

# Setting the labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show plot
plt.show()
```
